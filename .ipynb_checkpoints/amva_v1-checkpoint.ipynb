```python
dfincorrecto1 = dfincorrecto
```

```python
dfincorrecto1['PASAJEROSSUBEN'] =\
dfincorrecto1['PASAJEROSSUBEN'].str.replace('T', ' ')
```

```python
dfincorrecto1['PASAJEROSSUBEN'] =\
dfincorrecto1['PASAJEROSSUBEN'].str.replace('D', '')
```

```python
dfincorrecto1['PASAJEROSBAJAN'] =\
dfincorrecto1['PASAJEROSBAJAN'].str.replace('T', ' ')
```

```python
dfincorrecto1['PASAJEROSBAJAN'] =\
dfincorrecto1['PASAJEROSBAJAN'].str.replace('D', '')
```

```python
dfincorrecto1
```

```python
dfincorrecto1[['subendelantera','subentrasera']] = dfincorrecto1['PASAJEROSSUBEN'].loc[dfincorrecto1['PASAJEROSSUBEN'].str.split().str.len() == 2].str.split(expand=True)
dfincorrecto1
```

```python
dfincorrecto1[['bajandelantera','bajantrasera']] = dfincorrecto1['PASAJEROSBAJAN'].loc[dfincorrecto1['PASAJEROSBAJAN'].str.split().str.len() == 2].str.split(expand=True)
dfincorrecto1
```
